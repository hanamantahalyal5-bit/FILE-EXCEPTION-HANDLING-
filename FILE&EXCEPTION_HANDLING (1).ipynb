18) How would you implement a basic logging setup that logs to a file with rotation after 1M?

```python
import logging
from logging.handlers import RotatingFileHandler

# Step 1: Create a rotating file handler
handler = RotatingFileHandler(
    "fintech_app.log",    # Log file name
    maxBytes=1_000_000,   # Rotate after 1 MB
    backupCount=3         # Keep last 3 log files
)

# Step 2: Set format and level
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)

# Step 3: Create logger and attach handler
logger = logging.getLogger("fintech_logger")
logger.setLevel(logging.INFO)
logger.addHandler(handler)

# Step 4: Write test logs
for i in range(10000):
    logger.info(f"Transaction #{i} processed successfully.")
```

21) Write a Python program that reads a file and prints the number of occurrences of a specific word?

```python
# Ask user for file name and the word to search
filename = "my_one.txt"
word_to_search = "this"

# Step 1: Open the file safely using a context manager
with open(filename, "r") as file:
    content = file.read()

# Step 2: Count how many times the word appears
count = content.lower().count(word_to_search.lower())

# Step 3: Display result
print(f"The word '{word_to_search}' appears {count} times in {filename}.")
```
